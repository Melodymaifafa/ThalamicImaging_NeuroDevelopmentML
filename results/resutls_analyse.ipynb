```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Normalized Data Catogorical prediction

```python
cognitive_normalized_cat = pd.read_excel('results_normalized_cat_BIIICss.xlsx')
receptive_normalized_cat = pd.read_excel('results_normalized _cat_BIIIRss.xlsx')
expressive_normalized_cat = pd.read_excel('results_normalized _cat_BIIIEss.xlsx')
```

```python
cognitive_normalized_cat.drop(columns=['Feature Selection','Target'], inplace=True)
receptive_normalized_cat.drop(columns=['Feature Selection','Target'], inplace=True)
expressive_normalized_cat.drop(columns=['Feature Selection','Target'], inplace=True)
```

```python
cognitive_normalized_cat.set_index('Model',inplace=True)
receptive_normalized_cat.set_index('Model',inplace=True)
expressive_normalized_cat.set_index('Model',inplace=True)
```

```python
#BIIICss_normalized
plt.figure(figsize=(8, 4),dpi=200)
sns.heatmap(cognitive_normalized_cat, annot=True, fmt=".2f", cmap='Purples')
plt.title('Prediction of BayleyIII Cognitive Ability (normalized features)')
plt.show()
```

```python
#BIIIRss_normalized
plt.figure(figsize=(8, 4),dpi=200)
sns.heatmap(receptive_normalized_cat, annot=True, fmt=".2f", cmap='Purples')
plt.title('Prediction of BayleyIII Receptive Languge Ability (normalized features)')
plt.show()
```

```python
#BIIIEss_normalized
plt.figure(figsize=(8, 4),dpi=200)
sns.heatmap(expressive_normalized_cat, annot=True, fmt=".2f", cmap='Purples')
plt.title('Prediction of BayleyIII Expressive Language Ability (normalized features)')
plt.show()
```
